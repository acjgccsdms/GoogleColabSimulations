Doing a absorption experiement using a new random_walk function.

```python
from simlab.random_walk import random_walk_absorbing
import matplotlib.pyplot as plt

[path, hitting_time, boundary] = random_walk_absorbing(-150, 100)
plt.plot(path)
plt.title("Boundary Walk")
plt.show()
if hitting_time < 10000:
  print(f"The hitting time is: {hitting_time} and it hit the boundary at: {boundary}")
else: print(f"It never hits the boundary")
```

Giving stats for which boundary is hit.

```python
from simlab.random_walk import hitting_time
times, (prob_left, prob_right) = hitting_time(-40,10)
import matplotlib.pyplot as plt
import numpy as np

plt.plot(times)
plt.title("Hitting Times for Equal Boundaries of -20 and 20")
plt.show()
average_hit_time = np.mean(times)
print(average_hit_time)
print(f"The probability of hitting the left boundary is: {prob_left}\n and the probability of hitting the right boundary is: {prob_right}")

```
